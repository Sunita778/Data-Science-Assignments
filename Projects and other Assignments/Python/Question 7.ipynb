##### A. Get all the Earth meteorites that fell before the year 2000

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
import matplotlib.pyplot as plt
from datetime import datetime

n_df = pd.read_csv('nasa_resource.csv')

# Create a copy for better practice
ns_df = n_df.copy()

# Convert the 'timestamp' column to datetime data type using datetime
ns_df['year'] = ns_df['year'].apply(lambda x: datetime.strptime(x, "%Y-%m-%dT%H:%M:%S.%f") if pd.notnull(x) else None)

# Extract the year
ns_df['Year'] = ns_df['year'].apply(lambda x: x.year if x else None)

# Filter the DataFrame based on conditions
filtered_df = ns_df[(ns_df['fall'] == 'Fell') & (ns_df['Year'] < 2000)]

# Retrieve all the Earth meteorites that fell before the year 2000
names = filtered_df['name'].to_list()

names
```

```python
# Plot for the Earth meteorites that fell before the year 2000
import matplotlib.pyplot as plt

df = ns_df[(ns_df['fall'] == 'Fell') & (ns_df['Year'] < 2000)]

plt.figure(figsize=(40, 10))
plt.bar(df['name'], df['year'])
plt.xlabel('name')
plt.ylabel('year')
plt.title('the Earth meteorites that fell before the year 2000')
plt.xticks(rotation=90)
plt.show()
```

##### B. Get all the earth meteorites co-ordinates who fell before the year 1970

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
import matplotlib.pyplot as plt
from datetime import datetime

n_df = pd.read_csv('nasa_resource.csv')

# Creating a copy for better practice
ns_df = n_df.copy()

# Converting the 'timestamp' column to datetime data type using datetime
ns_df['year'] = ns_df['year'].apply(lambda x: datetime.strptime(x, "%Y-%m-%dT%H:%M:%S.%f") if pd.notnull(x) else None)

# Extract the year
ns_df['Year'] = ns_df['year'].apply(lambda x: x.year if x else None)

# Filtering based on conditions
filtered_df = ns_df[(ns_df['fall'] == 'Fell') & (ns_df['Year'] < 1970)]

# names of Earth meteorites that fell before the year 1970
names1 = filtered_df['name'].to_list()

names1
```

```python
# meteorites that fell and before the year 1970
filtered_df = ns_df[(ns_df['fall'] == 'Fell') & (ns_df['Year'] < 1970)]

# names of Earth meteorites that fell before the year 1970
names1 = filtered_df['name'].tolist()

# Plot for the count of meteorites
plt.figure(figsize=(40, 6))
plt.hist(names1, bins=len(names1), edgecolor='black')
plt.xlabel('Meteorite Name')
plt.ylabel('Count')
plt.title('Count of Earth Meteorites that Fell Before 1970')
plt.xticks(rotation=90)
plt.show()
```
