#####    A. Get all Pokemons whose spawn rate is less than 5%

```python
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
import matplotlib.pyplot as plt

#Read data
p_df = pd.read_excel('pokemon_pokedex.xlsx')

#Name of all Pokemons whose spawn rate is less than 5%(0.05 is for 5%)
p_df[p_df['spawn_chance']<0.05]['name'].to_list()
```

```python
# Plot for the spawn rate of the filtered Pokemon

filtered_pokemon = p_df[p_df['spawn_chance'] < 0.05]

plt.figure(figsize=(20, 6))
plt.bar(filtered_pokemon['name'], filtered_pokemon['spawn_chance'])
plt.xlabel('Pokemon')
plt.ylabel('Spawn Rate')
plt.title('Spawn Rate of Pokemon with Spawn Chance < 5%')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

##### B. Get all Pokemons that have less than 4 weaknesses

```python
import ast

p_df['weaknesses'] = p_df['weaknesses'].apply(ast.literal_eval)

#Create a new column for length of weaknesses
p_df['weak_lenghth'] = p_df['weaknesses'].apply(len)

#List of Pokemons that have less than 4 weaknesses
pok_name = p_df[p_df['weak_lenghth']<4]['name'].to_list()
pok_name
```

```python
import matplotlib.pyplot as plt

# Filter the Pokemon with less than 4 weaknesses
filtered_pokemon = p_df[p_df['weak_lenghth'] < 4]

# Plot the count of Pokemon with less than 4 weaknesses
plt.figure(figsize=(30, 10))
plt.hist(filtered_pokemon['name'], bins=len(filtered_pokemon), edgecolor='black')
plt.xlabel('Pokemon')
plt.ylabel('Weekness_Count')
plt.title('Count of Pokemon with Less than 4 Weaknesses')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

##### C. Get all Pokemons that have no multipliers at all

##### D. Get all Pokemons that do not have more than 2 evolutions

```python
p_df['next_evolution'] = p_df['next_evolution'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else {})

#Create a new column for length of weaknesses
p_df['evolution_len'] = p_df['next_evolution'].apply(len)

#List of Pokemons that have less than 4 weaknesses
pok_name = p_df[p_df['evolution_len']>2]['name'].to_list()

pok_name
```

##### E. Get all Pokemons whose spawn time is less than 300 seconds.

```python
# Extract minutes and seconds
p_df[['minutes', 'seconds']] = p_df['spawn_time'].str.split(':', expand=True)

# Convert minutes and seconds to numeric data types
p_df['minutes'] = pd.to_numeric(p_df['minutes'], errors='coerce')
p_df['seconds'] = pd.to_numeric(p_df['seconds'], errors='coerce')

# Replace NaN values with 0
p_df['minutes'] = p_df['minutes'].fillna(0)
p_df['seconds'] = p_df['seconds'].fillna(0)

# Convert minutes and seconds to integers
p_df['minutes'] = p_df['minutes'].astype(int)
p_df['seconds'] = p_df['seconds'].astype(int)

# Calculate total seconds
p_df['total_seconds'] = p_df['minutes'] * 60 + p_df['seconds']

# List of Pokemons whose spawn time is less than 300 seconds
p_name = p_df[p_df['total_seconds'] < 300].name.to_list()

p_name
```

```python
# Filter the Pokemon whose spawn time is less than 300 seconds
filtered_pokemon = p_df[p_df['total_seconds'] < 300]

# Plot the count of Pokemon with less than 4 weaknesses
plt.figure(figsize=(20, 6))
plt.hist(filtered_pokemon['name'], edgecolor='black')
plt.xlabel('Pokemon')
plt.ylabel('Spawn time')
plt.title('Count of Pokemon whose spawn time is less than 300 seconds')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
